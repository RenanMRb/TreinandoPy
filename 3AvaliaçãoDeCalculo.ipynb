```python
# Trabalho para terceira nota de "CÁLCULO DIFERENCIAL E INTEGRAL"
# Do almirante ao bacalhau, sou Vasco daqui à Portugal!
# 2ª Questão - Letra "A"

import sympy as sp
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Define a função
def func (x):
    return x ** 5

# Define o método do trapézio
def met_trap(a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n + 1)
    y = func (x)
    integral = h * (y[0] + 2 * np.sum(y[1:n]) + y[n]) / 2
    return integral, x, y

# Define os limites de integração
a = 1
b = 5

# Número de subintevalos para o método do trapézio (quanto maior os subintervalos, maior a precisão)
n = 100

# Cálcula a integral usando o método do trapézio
result, x_vals, y_vals = met_trap(a, b, n)

# Mostra o resultado da integral
print(f' O resultado da integral definida é aproximadamente: {result}')

# Gera e mostra o gráfico
plt.plot(x_vals, y_vals, color = 'orange', label = '$x^5$')
plt.fill_between(x_vals, y_vals, alpha = 0.2, color = 'yellow', label = 'Área Aproximada')
plt.title('Método do trapezio para $\int_{1}^{5} x^5 \, dx$', color = 'brown')
plt.xlabel('x', color = 'brown')
plt.ylabel('$x^5$', color = 'brown')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Trabalho para terceira nota de "CÁLCULO DIFERENCIAL E INTEGRAL"
# Se eu cantar é pra você. Se eu sorri é por você. Se eu torcer é por você: Vasco!
# 3ª Questão

import sympy as sp #Define as bibliotecas
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Define o método do trapézio
def met_trap(data):
    n = len(data) - 1
    delta_x = 1
    integ = (data[0][1] + data[n][1]) / 2
    for i in range(1, n):
        integ += data[i][1]
    integ = delta_x * integ
    return integ

# Define a regra de simpson
def reg_simp(data):
    n = len(data) - 1
    delta_x = 1
    integ = data[0][1] + data[n][1]
    for i in range(1, n, 2):
        integ += 4 * data[i][1]
    for i in range(2, n - 1, 2):
        integ += 2 * data[i][1]
    integ = (delta_x / 3) * integ
    return integ

# Tabela de dados
dados = [(9, 75.3),
         (10, 77.0),
         (11, 83.2),
         (12, 84.8),
         (13, 86.5),
         (14, 86.4),
         (15, 81.1),
         (16, 78.6)]

# Separa os dados em listas de tempo e temperatura para o gráfico
tempo, temperatura = zip(*dados)

# Calcula a integral usando o método do trapézio
int_result_trap = met_trap(dados)

# Calcula a integral usando a regra de Simpson
int_result_simp = reg_simp(dados)

# Cria funções de interpolação para os dados
interpt_lin = interp1d(tempo, temperatura, kind = 'linear')
interpt_cubic = interp1d(tempo, temperatura, kind = 'cubic')

# Gera um conjunto de pontos para as curvas interpoladas
tempo_interpt = np.linspace(min(tempo), max(tempo), 100)

# Cria subgráficos
fig, axs = plt.subplots(2, 1, figsize = (8, 10))

# cria e gera gráfico para o método do trapézio
axs[0].plot(tempo, temperatura, 'o', label = 'Dados de Temperatura', color = 'cyan')
axs[0].fill_between(tempo, temperatura, alpha = 0.2, label = 'Área Sob a Curva', color = 'yellow')
axs[0].plot(tempo_interpt, interpt_lin(tempo_interpt), label = 'Interpolação Linear', color = 'orange')
axs[0].plot(tempo_interpt, interpt_cubic(tempo_interpt), label = 'Interpolação Cúbica', color = 'lime')
axs[0].set_title('Método do Trapézio', color = 'brown')
axs[0].set_xlabel('Tempo (h)')
axs[0].set_xlabel('Temperatura (°F)', color = 'brown')
axs[0].legend()
axs[0].text(9, 85, f'Integral ≈ {int_result_trap:.2f} °F', bbox = dict(facecolor = 'white', alpha = 0.8), color = 'brown')

# cria e gera gráfico para o método de simpson
axs[1].plot(tempo, temperatura, 'o', label = 'Dados de Temperatura', color = 'cyan')
axs[1].fill_between(tempo, temperatura, alpha = 0.2, label = 'Área Sob a Curva', color = 'yellow')
axs[1].plot(tempo_interpt, interpt_lin(tempo_interpt), label = 'Interpolação Linear', color = 'orange')
axs[1].plot(tempo_interpt, interpt_cubic(tempo_interpt), label = 'Interpolação Cúbica', color = 'lime')
axs[1].set_title('Método do Trapézio', color = 'brown')
axs[1].set_xlabel('Tempo (h)')
axs[1].set_xlabel('Temperatura (°F)', color = 'brown')
axs[1].legend()
axs[1].text(9, 85, f'Integral ≈ {int_result_simp:.2f} °F', bbox = dict(facecolor = 'white', alpha = 0.8), color = 'brown')

# Ajustar layout
plt.tight_layout()

# Mostrar os subplots
plt.show()
```

```python
# Trabalho para terceira nota de "CÁLCULO DIFERENCIAL E INTEGRAL"
# Amo um time, vivo por uma torcida e sofro por um ideal: Vasco!
# 4ª Questão

import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from sympy import symbols, integrate as sympy_integrate, sqrt, pi

# Método analítico
y = sp.symbols('y')
a_int = sympy_integrate(sqrt(16 - y ** 2), (y, -4, 4))
rad = 4 # Raio da circuferência
a_analy = a_int * 4 / rad ** 2
pi_aprox_analy = a_analy.evalf()

# Método do trapézio
def f(y):
    return np.sqrt(16 - y ** 2)

def met_trap(f, a, b, n):
    h = (b - a) / n
    result = (f(a) + f(b)) / 2
    for i in range(1, n):
        result += f(a + i * h)
    result *= h
    return result

n_trap = 100 # Número de subintervalos
y_values = np.linspace(-4, 4, n_trap + 1)
a_num_trap = met_trap(f, -4, 4, n_trap)
pi_aprox_num_trap = a_num_trap * 4 / rad ** 2

# Gaussian quadrature com n = 4
n_gaus = 4
nodes, weights = np.polynomial.legendre.leggauss(n_gaus)
a_num_gaus = sum(weights * f(nodes))
pi_prox_num_gaus = a_num_gaus * 4 / rad ** 2

# Gera e mostra o gráfico
plt.figure(figsize = (8, 6))
plt.plot(y_values, f(y_values), label = 'Função $f(y)$', color = 'orange')
plt.fill_between(y_values, f(y_values), alpha = 0.3, label = 'Área sob a curva', color = 'yellow')
plt.title('Visualização da área sob a curva', color = 'brown')
plt.xlabel('$y$', color = 'brown')
plt.ylabel('$f(y)$', color = 'brown')
plt.legend()
plt.grid(True)
plt.show()

# Mostra resultados
print(f'Aproximação analítica de pi: {pi_aprox_analy}')
print(f'Aproximação numérica (método do trapézio) de pi: {pi_aprox_num_trap}')
print(f'Aproximação numérica (quadrature gaussian) de pi: {pi_prox_num_gaus}')
```
